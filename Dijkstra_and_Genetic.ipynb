#Node Class


The code segment below defines a class Node representing an object in the network, such as a car or base station. It is initialized with two key attributes: node_id is an identification number specifying which set of point coordinates belongs to any given node in the network. location stores its x and y coordinate values, its physical address. In particular, the location attribute is vital for spatial operations--in calculating distances between nodes and so on. These often form part of network analysis and optimization activities. This class thus constitutes an elemental segment in any system holding information distributed across spatially positioned nodes around a network.

```python
# Node Class for Vehicles and Base Stations
# This class represents the vehicles and base stations in the network.
class Node:
    def __init__(self, node_id, location):
        # Constructor for the Node class.
        # This class represents a node in the network, such as a vehicle or a base station.

        # node_id is a unique identifier for the node.
        # It's used to distinguish this node from others in the network.
        self.node_id = node_id

        # location represents the physical coordinates of the node.
        # This could be a tuple or a list containing the x and y coordinates.
        # It's important for calculating distances to other nodes.
        self.location = location
```

#Edge Class

The Edge class specifies the links or relationships (connections, edges between nodes) in a network. These include vehicles and base stations, for example. It has a constructor that initializes four key attributes: node1 and node2, identifiers for the two nodes which this edge connects; distance Which is actually an important thing because if you are going to go through all these calculations of signal strength or transmission delay then you have to know how far apart they are. And finally there's transmission rate representing data being transmitted.

```python
# Edge Class Representing Connections
# This class defines the edges or connections between the nodes (vehicles and base stations).
class Edge:
    def __init__(self, node1, node2, distance, transmission_rate):
        # Constructor for the Edge class.
        # This class represents an edge in the network, which is a connection between two nodes.

        # node1 and node2 are the identifiers (IDs) of the two nodes that this edge connects.
        self.node1 = node1
        self.node2 = node2

        # distance represents the physical distance between the two nodes.
        # This could be important for calculating signal strength or transmission delay.
        self.distance = distance

        # transmission_rate represents the rate at which data can be transmitted over this edge.
        # This could depend on various factors like distance, interference, etc.
        self.transmission_rate = transmission_rate
```

#Network Class

This class models a network consisting of nodes and edges. It includes methods for adding nodes and edges to the network, along with a static method for distance calculation:

**The __init__ method:** initializes the network with an empty dictionary for storing nodes and an empty list for edges.

**add_node Method:** Adds a new node to the network. The node is stored in the dictionary with its ID as the key.

**add_edge Method:** Creates an edge between two nodes. It checks if both nodes exist, calculates the distance between them, and adds the edge if the transmission rate is positive. The edge is represented by the Edge class, which is assumed to be defined elsewhere.

**calculate_distance Static Method**: Similar to the standalone function, this method calculates the distance between two locations. It's included in the class for encapsulation, making the class self-contained for operations involving distance calculations.

**transmission_rate Method**: Determines the data transmission rate based on the distance between nodes. It uses a series of conditional statements to assign lower transmission rates to longer distances, reflecting the typical behavior in real-world networks where signal strength diminishes with distance.

```python
import math

def calculate_distance(location1, location2):
    # This function calculates the Euclidean distance between two points.
    # The points are given as tuples (x, y).
    return math.sqrt((location1[0] - location2[0]) ** 2 + (location1[1] - location2[1]) ** 2)

class Network:
    def __init__(self):
        # Constructor for the Network class.
        # Initializes an empty dictionary for nodes and an empty list for edges.
        self.nodes = {}
        self.edges = []

    def add_node(self, node):
        # This method adds a node to the network.
        # The node is stored in the nodes dictionary, with the node's ID as the key.
        self.nodes[node.node_id] = node

    def add_edge(self, node1_id, node2_id, transmission_rate):
        # This method adds an edge between two nodes in the network.
        # It first checks if both nodes exist in the network.
        if node1_id in self.nodes and node2_id in self.nodes:
            # Calculate the distance between the two nodes.
            distance = calculate_distance(self.nodes[node1_id].location, self.nodes[node2_id].location)
            # If the transmission rate is positive, add the edge to the list of edges.
            if transmission_rate > 0:
                self.edges.append(Edge(node1_id, node2_id, distance, transmission_rate))


    def transmission_rate(self, distance):
        # This method determines the transmission rate based on the distance.
        # It uses a series of if-else conditions to return different rates.
        # The greater the distance, the lower the transmission rate.
        if distance >= 6000:
            return 0
        elif distance >= 4000:
            return 1
        elif distance >= 3000:
            return 2
        elif distance >= 2000:
            return 4
        elif distance >= 1000:
            return 6
        elif distance >= 500:
            return 8
        else:
            return 10
```

#Network Initialization Function


The initialize_network function establishes an integral network by generating a new Network object, and populating it with nodes representing vehicles and base stations. Starting with two congregated base stations which are already predefined, it then proceeds to go through the provided vehicle data (in DataFrame format) whose columns represent each number and coordinates of a given point for any vehicles one wishes to add. Each automobile is viewed as a node with its own identifier. Establishing nodes, function systematically forms edges between all node pairs; positive transmission rates are determined by their distances. The rate is computed via the Network class's transmission_rate method.

```python
# Function to Initialize the Network
# This function initializes the network with vehicles and base stations.
def initialize_network(vehicle_data):
    # Initialize a new Network object.
    network = Network()

    # We are adding two base stations to the network with predefined locations.
    network.add_node(Node("BS-1", (-1, -1)))
    network.add_node(Node("BS-2", (56325, 9)))

    # Iterate through the vehicle data to add each vehicle as a node in the network.
    # 'vehicle_data' is a DataFrame with columns 'No.', 'x', and 'y'.
    for index, row in vehicle_data.iterrows():
        # Create a node for each vehicle with a unique ID and location.
        network.add_node(Node(f"Car-{int(row['No.'])}", (row['x'], row['y'])))

    # Create edges between all pairs of nodes if they have a positive transmission rate.
    for node1_id in network.nodes:
        for node2_id in network.nodes:
            # Ensuring not to create an edge between a node and itself.
            if node1_id != node2_id:
                # Calculating the distance between the two nodes.
                distance = calculate_distance(network.nodes[node1_id].location, network.nodes[node2_id].location)
                # Determining the transmission rate based on the distance.
                trans_rate = network.transmission_rate(distance)
                # Adding an edge only if the transmission rate is positive.
                if trans_rate > 0:
                    network.add_edge(node1_id, node2_id, trans_rate)

    # Returning the fully initialized network.
    return network
```

#Latency Calculation Function

The calculate latency function calculates the total latency along a specific path throughout an entire network. It assumes that latency between each pair of connected nodes,links in the path is fixed and 50ms. In this implementation, delay_per-link is set to 50 milliseconds the time it takes for a signal to pass from any one node in the path through the next. The function returns the number of links in the path, which is one less than total count of all nodes. Because each link forms between two adjacent composed node elements. Multiplying this number of links by the constant delay per link gives us the total latency, or time needed for a signal to pass from source to destination.

```python
# Function to Calculate End-to-End Latency
# This function calculates the total latency of a given path in the network.
def calculate_latency(path):

    # delay_per_link is the fixed amount of time (in milliseconds) it takes for a signal
    # to travel from one node to another.
    delay_per_link = 50  # milliseconds

    # The number of links in the path is one less than the number of nodes in the path.
    # For example, a path with 3 nodes has 2 links.
    number_of_links = len(path) - 1

    # The total latency is calculated by multiplying the number of links with the delay per link.
    # This gives the total time taken for a signal to travel through the entire path.
    return number_of_links * delay_per_link
```

#Data Transmission Rate Calculation Function

The task of calculating what the lowest transmission rate is among all paths within a network must be done by the "calculate_data_transmission_rate" function, as it plays an important role in whether data bottlenecks exist on any path. It scrolls through node pairs which are next to one another in the path, and if able it finds an edge-connecting to this adjacent pair. The transmission rate of every edge discovered is added to a list. In other words, zero means a lack of an edge between two node pairs. The function returns the lowest rate from this list, which is still limiting for all of these pairs after processing paths.

```python
# Function to Calculate End-to-End Data Transmission Rate
# This function computes the minimum transmission rate along a given path.
def calculate_data_transmission_rate(path, network):
    # Initialize an empty list to store transmission rates of each edge in the path
    rates = []

    # Iterate over each pair of adjacent nodes in the path
    for i in range(len(path) - 1):
        # Get the IDs of the current pair of nodes
        node1_id = path[i]
        node2_id = path[i + 1]

        # Finding the edge in the network that connects these two nodes
        # This uses a generator expression to iterate over network.edges
        # and find the first edge where the nodes match the current pair
        edge = next((e for e in network.edges if (e.node1 == node1_id and e.node2 == node2_id)), None)

        # If such an edge is found, append its transmission rate to the rates list
        if edge:
            rates.append(edge.transmission_rate)
        else:
            # If no edge is found, append a transmission rate of 0
            rates.append(0)

    # Return the minimum transmission rate found in the path
    # If the rates list is empty, return 0
    return min(rates) if rates else 0
```

#Fitness Function

In the accompanying code, "fitness_function" evaluates how good a path is by using data transmission rate and latency to produce a score called fitness. For assessing transmission rate, it uses calculate_data_transmission_rate function; for latency calculation,it uses latency function. After this, the function gives weights to these factors: positive transmission rate being a plus and high latency considered negative. As a metric concerning the efficiency and effectivity of paths, this "fitness score" is established after adding weight to various factors. The specific weights are determined according processes for data transmission efficiency against time delays that give more consideration.

    weight_transmission_rate = 1.1
    weight_latency = 0.01
  
Here we have chosen weight as 1.1 to solve twin objective in a better way, it is going to create slight more preference on transmission rate side.

```python
# Fitness Function for Evaluating Paths
# This function evaluates the fitness of a path based on its data transmission rate and latency.
def fitness_function(path, network):
    # Calculate the data transmission rate for the given path in the network
    # This uses the function calculate_data_transmission_rate defined earlier
    transmission_rate = calculate_data_transmission_rate(path, network)

    # Calculate the latency for the given path
    # This assumes the existence of a function calculate_latency that is not provided here
    latency = calculate_latency(path)

    # Define the weights for transmission rate and latency
    # These weights determine how much each factor (transmission rate and latency) contributes to the fitness score
    weight_transmission_rate = 1.1
    weight_latency = 0.01

    # Calculate the fitness score
    # This is done by multiplying the transmission rate by its weight and subtracting the product of latency and its weight
    # A higher transmission rate contributes positively to the fitness, while higher latency contributes negatively
    fitness = (transmission_rate * weight_transmission_rate) - (latency * weight_latency)

    # Return the calculated fitness score
    return fitness
```

#Rank-Based Selection Function

Genetic algorithms' rank_based selection function sorts a population by fitness scores and selects individuals for the next generation according to their ranks. The higher ranked an individual is, the more likely it will be chosen. Its selection probabilities are inversely proportional to ranks. It chooses a subset according to these probabilities, and returns the selected result; fitter individuals are favored while diversity is maintained in population. For guiding the evolutionary process to optimal solutions, this method is all important.

```python
# Rank-Based Selection Function
# This function selects a subset of the population based on their fitness scores, using a rank-based approach.
def rank_based_selection(population, fitness_scores, selection_size):
    # Combine each individual in the population with its corresponding fitness score and sort them.
    # Sorting is done in descending order based on fitness scores, so higher fitness scores come first.
    ranked_population = sorted(zip(population, fitness_scores), key=lambda x: x[1], reverse=True)

    # Calculate the total of all rank positions. The rank of an individual is its position in the sorted list.
    total_ranks = sum(range(1, len(population) + 1))

    # Calculate the probability of selection for each individual based on its rank.
    # The probability is inversely proportional to the rank - higher ranked individuals have a higher probability.
    rank_probabilities = [rank / total_ranks for rank in range(len(population), 0, -1)]

    # Prepare a list to store the selected individuals.
    selected_parents = []

    # Select individuals to form the new population.
    for _ in range(selection_size):
        # Randomly select an index based on the calculated rank probabilities.
        selected_index = np.random.choice(len(population), p=rank_probabilities)

        # Append the selected individual (not the fitness score) to the list of selected parents.
        selected_parents.append(ranked_population[selected_index][0])

    # Return the selected subset of individuals.
    return selected_parents
```

#Population Initialization Function

The initialize_population function populates an initial population of paths for a genetic algorithm, in the case of a network with vehicles and base stations. It begins by selecting all vehicle nodes, each of whose IDs starts with a "Car-", and another group of base station nodes with an ID including BS-. The function then creates a given number of paths, each made up of random permutations of the vehicle nodes followed by one randomly selected base station. Such a structure is based on the premise that each path in network should terminate at some base station. These paths become the initial population for the genetic algorithm.

```python
# Function to Initialize Population
# This function creates an initial population of paths for the genetic algorithm.
def initialize_population(network, population_size):
    # Initialize an empty list to store the population of paths.
    population = []

    # Extract all nodes representing vehicles from the network.
    # These are identified by their IDs starting with "Car-".
    vehicle_nodes = [node_id for node_id in network.nodes if node_id.startswith("Car-")]

    # Extract all nodes representing base stations from the network.
    # These are identified by their IDs starting with "BS-".
    base_stations = [node_id for node_id in network.nodes if node_id.startswith("BS-")]

    # Generate a specified number of paths to form the initial population.
    for _ in range(population_size):
        # Copy the list of vehicle nodes to create a new path.
        path = vehicle_nodes.copy()

        # Randomly shuffle the order of the vehicles in the path.
        random.shuffle(path)

        # Randomly select a base station and append it to the end of the path.
        # The assumption here is that each path should end at a base station.
        path.append(random.choice(base_stations))

        # Add the generated path to the population.
        population.append(path)

    # Return the populated list of paths.
    return population
```

#Population Initialization for Specific Vehicle Function



Initialize_population It finds nearby nodes to form paths, with these paths being unique and possibly terminating at some base station. These paths are generated iteratively, until the population size is reached. This approach selects the specific vehicle in question to optimize routes.

```python
# Function to Initialize Population for a Specific Vehicle
# This function creates a population of paths starting from a specific vehicle node.
# Function to Initialize Population for a Specific Vehicle
def initialize_population_for_vehicle(network, vehicle_node, population_size):

    # Initialize an empty list to store the population of paths.
    population = []

    # Get a list of all nodes in the network.
    all_nodes = list(network.nodes.keys())

    # Extract all nodes representing base stations from the network.
    base_stations = [node_id for node_id in network.nodes if node_id.startswith("BS-")]

    # Define a function to find nodes close to a given node, excluding specified nodes.
    def find_close_nodes(node, exclude):
        node_location = network.nodes[node].location
        # Return nodes that are not the given node, not in the exclude list, and within 6000 units of distance.
        return [n for n in all_nodes if n != node and n not in exclude and calculate_distance(node_location, network.nodes[n].location) < 6000]

    # Define a function to generate a path starting from a given node.
    def generate_path(start_node):
        path = [start_node]
        current_node = start_node

        # Continuously extend the path with close nodes.
        while True:
            # Find nodes close to the current node, excluding those already in the path and base stations.
            close_nodes = find_close_nodes(current_node, path + base_stations)

            if not close_nodes:
                # If no close nodes are found, end the path.
                break

            next_node = random.choice(close_nodes)
            path.append(next_node)
            current_node = next_node

            # Randomly decide whether to end the path at a base station.
            if len(path) > 2 and random.choice([True, False]):
                final_station = random.choice(base_stations)
                if calculate_distance(network.nodes[current_node].location, network.nodes[final_station].location) < 6000:
                    path.append(final_station)
                    break

        return path

    # Generate paths until the population reaches the desired size.
    while len(population) < population_size:
        new_path = generate_path(vehicle_node)

        # Add the new path to the population if it's unique and has more than just the start node.
        if new_path not in population and len(new_path) > 1:
            population.append(new_path)

    # Return the generated population of paths.
    return population


```

#Ordered Crossover Function

ordered_crossover performs the crossover operation, which is often used in genetic algorithms. It starts by taking on the first node of parent1. Then it tries to add successively all those other nodes of parent 1 which are not already part of this branch and still meet a given distance criterion. Hence because of this restriction, if node from parent1 doesn't do it then the function looks for one among everything else that hasn't been added in parent2. Moreover, the operation includes that all offspring paths terminate at a base station. This is an absolutely necessary condition in the network environment. If it isn't a base station, the last node finds an amenable base station that terminates its path. This provides a way for an offspring to inherit characteristics from each parent, while still observing the limitations of networks. This is an important consideration in a genetic algorithm for evolving and optimizing network paths.

```python

# This function performs the ordered crossover operation between two parent paths.
# Ordered Crossover Function
def ordered_crossover(parent1, parent2):
    # Define a function to check if a node can be added to the current path.
    # This checks the distance constraint between the last node in the path and the new node.
    def can_add_to_path(path, node):
        # Ensure both nodes are part of the network before calculating the distance.
        if path[-1] in network.nodes and node in network.nodes:
            return calculate_distance(network.nodes[path[-1]].location, network.nodes[node].location) < 6000
        return False

    # Start the offspring path with the first node of parent1.
    offspring = [parent1[0]]

    # Iterate through the nodes in parent1 and try to add them to the offspring path.
    for i in range(1, len(parent1)):
        # Add the next node from parent1 to the offspring if it meets the distance constraint.
        if can_add_to_path(offspring, parent1[i]):
            offspring.append(parent1[i])
        else:
            # If the node from parent1 doesn't meet the constraint, look for a valid node in parent2.
            for node in parent2:
                # Check if the node from parent2 is valid and not already in the offspring path.
                if can_add_to_path(offspring, node) and node not in offspring:
                    offspring.append(node)
                    break

    # Ensure the offspring path ends with a base station.
    # If the last node is not a base station, search for a base station that can be added.
    if not offspring[-1].startswith("BS-"):
        for bs in base_stations:
            if can_add_to_path(offspring, bs):
                offspring.append(bs)
                break

    # Return the generated offspring path.
    return offspring

```

#Generic Algorithm Loop


**Genetic Algorithms (GA)**:

Genetic algorithms, or GAs, are strong optimization techniques derived from natural selection theory. GA was first proposed by John Holland in the 1970s. It is modelled after natural selection, with solutions acting as chromosomes that are subject to genetic processes such as crossover, mutation, and selection. Subsequent generations are produced by applying these processes to a random population. An individual's odds of becoming a parent in the reproductive process are influenced by its fitness, which reflects the quality of the solution. GA has applications in machine learning, research and development, and optimization. It is skilled at handling computationally difficult issues(Ochelska-Mierzejewska et al. 2021).

A genetic algorithm works by first creating a population of solutions that are different from one another, and then applying genetic operations repeatedly. Over the course of several generations, the algorithm seeks to enhance the quality of solutions until a desirable result is obtained. The idea of a random search strategy proposed by Charles Darwin aids in the investigation of a predetermined search field. In-depth discussions of genetic algorithms' basic principles, process, and applications across a range of industries are covered in this paper (Fisher 1994).


This main execution block employs a genetic algorithm to determine the optimal network paths for one designated vehicle. It sets the algorithm's parameters, initializes the network and then attempts to optimize paths from a chosen vehicle node. Every generation the algorithm evaluates path fitness, selects parent paths and through crossover generates new paths. The cycle repeats, driving the population toward optimal solutions. Finally, it determines and shows the best route for the car. This includes such parameters as transmission rate and latency.



```python


# Initialize the network with vehicle data
network = initialize_network(vehicle_data)

# Set parameters for the genetic algorithm
population_size = 500  # Adjust as needed
number_of_generations = 20  # Number of generations

# Choose a specific vehicle to track
vehicle_to_track = 41 # Replace with the desired vehicle number
vehicle_node = f"Car-{vehicle_to_track}"

# Initialize population for the chosen vehicle in the network
population = initialize_population_for_vehicle(network, vehicle_node, population_size)

# Iterate through each generation
for generation in range(number_of_generations):
    # Calculate fitness scores for each path in the population
    fitness_scores = [fitness_function(path, network) for path in population]

    # Uncomment to print generation details
    #print(f"Generation {generation + 1} for Vehicle {vehicle_node}:")

    # Uncomment to print details of each path in the generation
    #for path in population:
    #    transmission_rate = calculate_data_transmission_rate(path, network)
    #    latency = calculate_latency(path)
    #    fitness = fitness_function(path, network)
    #    #print(f"  Path: {path}, Transmission Rate: {transmission_rate}, Latency: {latency}ms, Fitness Score: {fitness}")

    # Perform selection to choose parents for the next generation
    parents = rank_based_selection(population, fitness_scores, len(population))

    # Create the next generation using crossover
    offspring = []
    for i in range(0, len(parents), 2):
        if i + 1 < len(parents):
            # Perform ordered crossover between pairs of parents
            crossover_result = ordered_crossover(parents[i], parents[i + 1])

            # Handle the results of the crossover
            if isinstance(crossover_result, tuple):
                # If no valid crossover was possible, use parents as they are
                offspring.extend(crossover_result)
            else:
                # Add the successfully generated offspring
                offspring.extend([crossover_result])

    # Update the population for the next generation, maintaining the population size
    population = offspring + population[:population_size - len(offspring)]

# Select the best path for the chosen vehicle based on fitness scores
best_path = max(population, key=lambda path: fitness_function(path, network))

# Calculate transmission rate and latency for the best path
best_transmission_rate = calculate_data_transmission_rate(best_path, network)
best_latency = calculate_latency(best_path)

# Function to construct the routing path string with transmission rates
def construct_routing_path_string(path, network):
    routing_path = []
    for i in range(len(path) - 1):
        node1_id = path[i]
        node2_id = path[i + 1]
        edge = next((e for e in network.edges if (e.node1 == node1_id and e.node2 == node2_id)), None)
        rate = edge.transmission_rate if edge else 0
        routing_path.append(f"({node2_id}, {rate})")
    return ', '.join(routing_path)

# Construct the routing path string
routing_path_str = construct_routing_path_string(best_path, network)

# Construct the output dictionary
output = {
    "Source Node": best_path[0],
    "Routing Path": routing_path_str,
    "End-to-end data rate": best_transmission_rate
}

# Print the output in the desired format
print(output)
```

#Dijkstra Algorithm

Dijkstra's algorithm solves the single-source shortest routes problem in a graph and is called for its discoverer, E.W. Dijkstra. Remarkably, this technique can compute the shortest pathways to every point in a graph at the same time. It does this by efficiently determining the shortest path from a source point to a destination. The efficacy of the algorithm stems from its capacity to methodically investigate and assess pathways, choosing the best option at each stage. Dijkstra's algorithm's repetitive process of determining and updating the shortest path to each network vertex forms the basis of the algorithm. The algorithm keeps track of the vertices that have known shortest paths, and it keeps growing this collection until the destination is reached. Applications for Dijkstra's algorithm is numerous in the disciplines of network routing, transportation systems, and other areas where shortest path finding is crucial(Gayathri and Chandrakala). To sum up, Dijkstra's algorithm and genetic algorithms are both effective methods for pathfinding and optimization, respectively. While Dijkstra's Algorithm effectively resolves the shortest routes problem in graph theory, Genetic Algorithms use natural evolution as inspiration to address complicated problems. Gaining an understanding of the fundamentals and practical uses of these algorithms is essential to maximizing their potential for solving problems across a variety of fields.

Here's the code defining several functions to calculate rates of transmission, implement modified Dijkstra ' s algorithm and find least-cost paths for a network model comprising vehicles and base stations: It computes inter-pair transfer rates according to distances, modifies Dijkstra's algorithm for constant latency per link connection and expands the network with base stations. The code also recalculates the best paths for each vehicle to the nearest base station, based on both latency and transmission rates.

```python
# Function to calculate distance between two points
def calculate_distance(x1, y1, x2, y2):
    return np.sqrt((x2 - x1)**2 + (y2 - y1)**2)

# Function to determine transmission rate based on distance
def transmission_rate(distance):
    if distance >= 6000:
        return 0
    elif distance >= 4000:
        return 1
    elif distance >= 3000:
        return 2
    elif distance >= 2000:
        return 4
    elif distance >= 1000:
        return 6
    elif distance >= 500:
        return 8
    else:
        return 10

data=vehicle_data
# Calculating pairwise transmission rates
num_vehicles = len(data)
rates = np.zeros((num_vehicles, num_vehicles))

for i in range(num_vehicles):
    for j in range(num_vehicles):
        if i != j:
            dist = calculate_distance(data.iloc[i]['x'], data.iloc[i]['y'], data.iloc[j]['x'], data.iloc[j]['y'])
            rates[i, j] = transmission_rate(dist)

# Adjusted function for modified Dijkstra's algorithm with fixed latency per link
def adjusted_dijkstra(start_node, num_nodes, rates):
    fixed_latency = 50  # in milliseconds
    max_distance = float('inf')
    min_rate = 0
    distance_to = [max_distance] * num_nodes
    rate_to = [min_rate] * num_nodes
    distance_to[start_node] = 0
    rate_to[start_node] = float('inf')  # Start with the highest possible rate

    previous_node = [-1] * num_nodes
    pq = PriorityQueue()
    pq.put((0, start_node))

    while not pq.empty():
        current_distance, current_node = pq.get()

        for neighbor in range(num_nodes):
            if rates[current_node][neighbor] > 0:  # Check if neighbor is connected
                new_distance = current_distance + fixed_latency
                new_rate = min(rate_to[current_node], rates[current_node][neighbor])

                if new_distance < distance_to[neighbor] or (new_distance == distance_to[neighbor] and new_rate > rate_to[neighbor]):
                    distance_to[neighbor] = new_distance
                    rate_to[neighbor] = new_rate
                    previous_node[neighbor] = current_node
                    pq.put((new_distance, neighbor))

    return distance_to, rate_to, previous_node

# Adding base stations as special nodes in the graph
BS_1 = (-1, -1)
BS_2 = (56325, 9)
base_stations = [BS_1, BS_2]

# Extend the rates matrix to include base stations
extended_rates = np.zeros((num_vehicles + len(base_stations), num_vehicles + len(base_stations)))
extended_rates[:num_vehicles, :num_vehicles] = rates

for i, bs in enumerate(base_stations):
    for j in range(num_vehicles):
        dist_to_bs = calculate_distance(data.iloc[j]['x'], data.iloc[j]['y'], bs[0], bs[1])
        rate_to_bs = transmission_rate(dist_to_bs)
        extended_rates[j, num_vehicles + i] = rate_to_bs
        extended_rates[num_vehicles + i, j] = rate_to_bs

# Function to reconstruct the path from the previous_node array
def reconstruct_path(previous_node, start_node, end_node):
    path = []
    current_node = end_node
    while current_node != start_node:
        path.append(current_node)
        current_node = previous_node[current_node]
    path.append(start_node)
    path.reverse()
    return path


# Updated function to find optimal paths and reconstruct the paths for each vehicle to both base stations
def find_and_reconstruct_optimal_paths(num_vehicles, extended_rates, data):
    optimal_paths = {}

    for vehicle in range(num_vehicles):
        distance_to_BS1, rate_to_BS1, prev_node_to_BS1 = adjusted_dijkstra(vehicle, num_vehicles + len(base_stations), extended_rates)
        distance_to_BS2, rate_to_BS2, prev_node_to_BS2 = adjusted_dijkstra(vehicle, num_vehicles + len(base_stations), extended_rates)

        # Choose the base station with the shortest path
        if distance_to_BS1[num_vehicles] <= distance_to_BS2[num_vehicles + 1]:
            path_to_BS = reconstruct_path(prev_node_to_BS1, vehicle, num_vehicles)
            end_to_end_rate = rate_to_BS1[num_vehicles]
            bs_label = "BS-1"
        else:
            path_to_BS = reconstruct_path(prev_node_to_BS2, vehicle, num_vehicles + 1)
            end_to_end_rate = rate_to_BS2[num_vehicles + 1]
            bs_label = "BS-2"

        # Format the routing path
        routing_path = []
        for i in range(1, len(path_to_BS)):  # Start from 1 to exclude the source node
            node = path_to_BS[i]
            prev_node = path_to_BS[i - 1]
            if node < num_vehicles:
                rate = rates[prev_node][node]
                routing_path.append(f"(Car-{node + 1}, {rate})")

        # Add the base station to the end of the path
        routing_path.append(f"({bs_label}, {end_to_end_rate})")

        optimal_paths[f"Car-{vehicle + 1}"] = {
            "Source Node": f"Car-{vehicle + 1}",
            "Routing Path": ', '.join(routing_path),
            "End-to-end data rate": end_to_end_rate
        }

    return optimal_paths

# Finding and reconstructing the optimal paths for all vehicles
optimal_paths_with_reconstruction = find_and_reconstruct_optimal_paths(num_vehicles, extended_rates, data)

# Print the results in the desired format
for key, value in optimal_paths_with_reconstruction.items():
    print(value)
```
